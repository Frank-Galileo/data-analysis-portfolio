## 4. Data Cleaning
🇬🇧 **English Version**\
Before performing the analysis, the dataset required several cleaning and preparation steps to ensure data consistency and usability.

Because the dataset is composed of multiple relational tables, the preparation process involved handling missing values, converting data types, and merging datasets into a unified analytical table.

### Handling Missing Values
Several columns contained missing values, particularly those related to delivery and review information. These missing values were expected because not all orders reach the same stages of the order lifecycle (for example, canceled orders may not have delivery timestamps).

For the purposes of this analysis:

- Missing delivery dates were left unchanged when they represented incomplete orders.
- Review data was merged when available to analyze customer satisfaction.

### Converting Data Types
Date columns were converted to proper datetime format to allow time-based analysis.

The following columns were converted:

- order_purchase_timestamp  
- order_approved_at  
- order_delivered_carrier_date  
- order_delivered_customer_date  
- order_estimated_delivery_date  

This conversion enables operations such as calculating delivery delays and analyzing monthly trends.

### Creating Derived Variables
Additional variables were created to support the analysis.

**Revenue**
Revenue was calculated by combining the product price and freight value for each order item:

Revenue = price + freight_value

This allows the estimation of total marketplace revenue.

**Delivery Delay**
Delivery delay was calculated as the difference between the actual delivery date and the estimated delivery date.

Delivery Delay = actual_delivery_date − estimated_delivery_date

Negative values indicate that the order was delivered earlier than expected.

### Building the Analytical Dataset
To perform the analysis, several tables were merged into a single dataset called `sales`. This dataset combines information from:

- orders
- order items
- customers
- products
- reviews

By consolidating these datasets, it becomes possible to analyze marketplace performance across multiple dimensions such as time, product categories, geography, and seller activity.

## 6. Insights and Findings
🇬🇧 **English Version**\
The exploratory analysis reveals several important insights about the structure and performance of the Brazilian e-commerce marketplace.

### Marketplace Growth
The number of orders increased significantly between 2017 and 2018, indicating strong marketplace growth during this period. This suggests increasing platform adoption and rising customer demand.

### Revenue Performance
Total marketplace revenue reached approximately **15.8 million BRL** across nearly **100,000 orders**. The **average order value is approximately 161 BRL**, indicating that revenue growth is primarily driven by an increase in order volume rather than higher spending per transaction.

### Product Category Performance
Several product categories dominate marketplace revenue. The top-performing categories include:

- Health & Beauty  
- Watches & Gifts  
- Bed, Bath & Table  
- Sports & Leisure  
- Computer Accessories  

These categories represent everyday consumer goods and popular gift items, which tend to generate consistent purchasing behavior.

### Customer Satisfaction
Customer reviews indicate a high level of satisfaction on the platform. More than **75% of review scores are rated 4 or 5 stars**, suggesting that the majority of customers are satisfied with their purchasing experience.

### Delivery Performance
Delivery performance appears to have a measurable impact on customer satisfaction. Orders associated with higher review scores tend to be delivered earlier relative to the estimated delivery date, suggesting that faster deliveries contribute to higher customer ratings.

### Geographic Distribution
Marketplace revenue is highly concentrated in economically developed regions of Brazil. The state of **São Paulo generates approximately 37% of total revenue**, followed by other major states such as Rio de Janeiro and Minas Gerais.

### Seller Ecosystem
The marketplace includes approximately **3,095 active sellers**, indicating a diverse seller ecosystem. The top seller generates roughly **1.6% of total marketplace revenue**, suggesting that the platform is not overly dependent on a single seller and that revenue is distributed across many participants.

- la distribution du revenu est-elle équilibrée ?

```python
# Concentration of income by seller. The order_items table already contains seller_id,
# and in most cases we already have the data in sales. Let's check:

# Concentration du revenu par vendeur. La table order_items contient déjà seller_id,
# dans la plupart des cas nous avons déjà les données dans sales. Vérifions :

"seller_id" in sales.columns
```

```python
# Calculation of income per salesperson
# Calcul du revenu par vendeur

revenue_by_seller = sales.groupby("seller_id")["revenue"].sum().sort_values(ascending=False)

revenue_by_seller.head(10)
```

```python
# How many sellers are there in the marketplace?
# Combien y a-t-il de vendeur dans la marketplace ?

sales["seller_id"].nunique()
```

```python
# Calculation of income concentration
# Calcul de la concentration du revenu

seller_revenue_df = revenue_by_seller.reset_index()
seller_revenue_df.columns = ["seller_id", "revenue"]
```

```python
# Sorts sellers
# Trie des vendeurs

seller_revenue_df = seller_revenue_df.sort_values("revenue", ascending=False)
```

```python
# Calculation of cumulative income
# Calcul du revenu cumulatif

seller_revenue_df["cumulative_revenue"] = seller_revenue_df["revenue"].cumsum()
seller_revenue_df["cumulative_percentage"] = seller_revenue_df["cumulative_revenue"] / seller_revenue_df["revenue"].sum()
```

```python
# Revenue Concentration by Sellers

seller_revenue_df["seller_rank"] = range(1, len(seller_revenue_df) + 1)
seller_revenue_df["seller_percentage"] = seller_revenue_df["seller_rank"] / len(seller_revenue_df)
```

# **Chart (Pareto / 80-20 rule)**

```python
plt.figure(figsize=(8,6))

plt.plot(
    seller_revenue_df["seller_percentage"],
    seller_revenue_df["cumulative_percentage"]
)

plt.title("Revenue Concentration by Sellers")
plt.xlabel("Percentage of Sellers")
plt.ylabel("Cumulative Revenue")

plt.show()
```
